```python
import pandas as pd 
```

```python
dados = pd.read_csv('sales_data.csv')
dados.head()
```

```python
dados = dados.drop(columns=['Day','Month','Year','Customer_Age','State'])
dados.head()
```

```python
dados.info()
```

```python
dados = dados.rename (columns={
    'Order_Quantity ':'quantidade',
    'Unit_Cost':'custo_unidade',
    'Unit_Price':'preço_unidade',
    'Profit':'lucro',
    'Cost':'custo',
    'Revenue':'revenda'
})
```

```python
dados.head()
```

```python
dados.info()
```

```python
dados = dados.rename(columns={'Order_Quantity':'quantidade'})
```

```python
dados.head()
```

```python
dados['revenda'] = dados['preço_unidade'] * dados['quantidade']
dados.head()
```

```python
dados['custo'] = dados['quantidade'] * dados['custo_unidade']
dados['lucro'] = dados['revenda'] - dados['custo']
dados.head()
```
